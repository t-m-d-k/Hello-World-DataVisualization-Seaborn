```python
ruralData = inputData[['Name of the District','Number of ICCC in Rural Areas as on 20.05.2021','Number of Proposed Beds under ICCC in Rural Areas as on 20.05.2021','Total Beds available in Rural Areas as on 20.05.2021','Beds occupied in Rural Areas as on 20.05.2021']]
fomratedData = pd.melt(ruralData, id_vars="Name of the District", var_name="Category", value_name="Bed Availability")
fomratedData
```

```python
pt.figure(figsize=[15.0,15.0],dpi=120,frameon=True)
#sns.barplot(x='Name of the District', y='Bed Availability', hue='Category', data=fomratedData)
ax = sns.barplot(x='Name of the District', y='Bed Availability', hue='Category', data=fomratedData)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
pt.tight_layout()
pt.show()

```

####  Rural district Conclusion :
##### district like salem , ranipet , sivaganga , villupuram there is not active beds available.which denotes government has alloted more beds in these districts

### URBAN area analysis

```python
UrbanData = inputData[['Name of the District','Number of Proposed ICCC in Urban Areas as on 20.05.2021','Number of Proposed Beds under ICCC in Urban Areas as on 20.05.2021']]
fomratedData = pd.melt(UrbanData, id_vars="Name of the District", var_name="Category", value_name="Bed Availability")
fomratedData
```

```python
pt.figure(figsize=[15.0,15.0],dpi=120,frameon=True)
#sns.barplot(x='Name of the District', y='Bed Availability', hue='Category', data=fomratedData)
ax = sns.barplot(x='Name of the District', y='Bed Availability', hue='Category', data=fomratedData)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
pt.tight_layout()
pt.show()

```

### Overall Conclusion :
### Could notice number of proposed bed in rural area is more then the urban area
### As the data set has no much information on the urban area , from the given data we can conclude number of beds proposed is comparitively high in salem 

### Now lets do some data visualization techniques in seaborn to understand the data more 

####  Pairplot , is the most efficient tool to plot all the data in the dataset and bring out their overal relationship between the data, shown below is the pair plot of the full dataset

```python
renamedInput = inputData.rename(columns = {'Number of ICCC in Rural Areas as on 20.05.2021': 'ICCC in Rural', 'Number of Proposed Beds under ICCC in Rural Areas as on 20.05.2021': 'Rural Proposed Beds','Total Beds available in Rural Areas as on 20.05.2021': 'Rural Total Beds' , 'Beds occupied in Rural Areas as on 20.05.2021': 'Rural Beds Occupied', 'Number of Proposed ICCC in Urban Areas as on 20.05.2021' : 'ICC urban area', 'Number of Proposed Beds under ICCC in Urban Areas as on 20.05.2021' : 'Urban Proposed beds'}, inplace = False)
renamedInput.head()
```

```python
g = sns.PairGrid(renamedInput, hue = 'Name of the District')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()

```
